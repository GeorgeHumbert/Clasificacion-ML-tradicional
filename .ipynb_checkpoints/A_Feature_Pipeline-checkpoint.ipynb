```python
import pandas as pd

# Cargar el dataset Titanic desde el archivo CSV
df = pd.read_csv('data/titanic.csv')

# Mostrar las primeras filas para verificar la carga
df.head()
```

```python
# Crear la columna FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Mostrar las primeras filas para verificar la nueva columna
df[['SibSp', 'Parch', 'FamilySize']].head()
```

```python
# Rellenar los valores nulos en la columna Age con la media
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Rellenar los valores nulos en Embarked con la moda
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Verificar si los valores nulos han sido eliminados
df.isnull().sum()
```

```python
# Codificar la variable Sex
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Usar get_dummies para Embarked
df = pd.get_dummies(df, columns=['Embarked'])

# Mostrar las primeras filas para verificar las transformaciones
df.head()
```
